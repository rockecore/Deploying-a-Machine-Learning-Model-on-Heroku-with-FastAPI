```python
import pandas as pd

data = pd.read_csv(r"...\starter\data/census.csv")
data
```

```python
mycols = list(data.columns)
mycols
```

```python
new_cols = [x.replace(" ", "") for x in mycols]    
new_cols = [x.replace("-", "_") for x in new_cols]
new_cols
```

```python
data.columns = new_cols
data.info()
```

```python
data.isnull().sum()
```

```python

pd.Index(data.workclass).is_object()
```

```python
## instructions say to remove spaces in the data too. I'm also going to get rid of those nasty dashes

def remove_spaces(col):
    if pd.Index(col).is_object():
        col = [x.replace(" ", "") for x in list(col)]
        col = [x.replace("-", "_") for x in col]
    return col

clean_data = data.apply(
    remove_spaces, axis=0)

clean_df = pd.DataFrame.from_dict(
    dict(zip(clean_data.index, clean_data.values))).transpose()
clean_df.columns = new_cols
clean_df
```

```python
## I think that is good
clean_df.to_csv(r"...\starter\data/clean_census.csv", index=False)
```
